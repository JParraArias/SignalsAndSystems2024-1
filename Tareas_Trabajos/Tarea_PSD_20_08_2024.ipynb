Teniendo en cuenta el Power Spectral Density (PSD), hallar la solución de:
$A \cdot \text{rect}_\tau(t)$ aplicando el método de welch.scipy con hanning window.

Simulación:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
from scipy.signal.windows import hann

# Parámetros
A = 2  # amplitud
tau = 0.5  # duración del pulso
fs = 2000  #  frecuencia de muestreo
t = np.arange(-5*tau, 5*tau, 1/fs)  #  rango de tiempo

# Se Define la señal rectángulo con los parámetros
x_t = A * np.where(np.abs(t) <= tau/2, 1, 0)

# Aplicar el método de Welch con ventana de Hanning
window = hann(len(x_t))
f, Pxx = welch(x_t, fs, window=window, nperseg=len(x_t))

# Graficar la Densidad Espectral de Potencia (PSD)
plt.figure(figsize=(10, 6))
plt.semilogy(f, Pxx)
plt.title('Densidad Espectral de Potencia (PSD) usando el Método de Welch')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('PSD [V^2/Hz]')
plt.grid(True)
plt.show()
```
